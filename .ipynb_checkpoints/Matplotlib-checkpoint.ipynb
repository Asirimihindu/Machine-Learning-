# Data Visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
#pyplot API
plt.plot();
```

```python
X=[1,2,3,4]
Y=[11,21,13,40]
plt.plot(X,Y);
```

```python
fig=plt.figure()
ax=fig.add_subplot()
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(X,Y)
plt.show()
```

```python
#recommended method Object-Oriented API
fig, ax =plt.subplots()
ax.plot(X,Y);
```

```python
x=[1,2,3,4]
y=[11,22,33,44]

fig, ax=plt.subplots(figsize=(10, 10))
ax.plot(x,y)
ax.set(title="Simple Plot" ,xlabel="x-axis",ylabel="y-axis");
fig.savefig("sample_plot.png")
```

```python
x=np.linspace(0,10,100)
fig,ax=plt.subplots()
ax.plot(x,x**3);
```

```python
x=np.linspace(0,10,100)
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x)**2);
```

```python
x=np.linspace(0,10,100)
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x)*3);
```

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
plt.style.use('seaborn-whitegrid')
fig, ((ax1,ax2),(ax3,ax4)) =plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x, x/2,label='x/2');
ax1.plot(-1,-1,"r+");
ax1.plot(x,x/3,label='x/3');
ax1.plot(x,x/4,label='x/4');
ax1.legend(loc='lower right', shadow=False, fontsize='x-small')
ax2.scatter(x,np.sin(x)*3);
ax3.hist(np.random.randn(1000));
ax4.scatter(x,np.exp(x)**2);
```

# Plotting From pandas DataFrame

```python
car_sales=pd.read_csv("car_sales.csv")
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales Date"]=pd.date_range("1/1/2020",periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="Sales Date",y="Total Sales")
```

```python
fig, ax=plt.subplots(figsize=(10, 10))
sct=ax.plot(car_sales["Sales Date"].astype(str).str[5:10],car_sales["Total Sales"])
ax.set(title="Simple Plot" ,xlabel="Sales Date",ylabel="Total Sales");
ax.legend(["Total Sales"]);
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
over_50=heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True);
sct=ax0.scatter(over_50["age"],over_50["chol"],c=over_50["target"],cmap="winter");
ax0.set(title="Heart Disease & Cholesterol Level",ylabel="Cholesterol");
ax0.legend(*sct.legend_elements(),title="Target");
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

ax0.set_xlim([50,80])

sct=ax1.scatter(over_50["age"],over_50["thalach"],c=over_50["target"],cmap="winter");
ax1.set(title="Heart Disease & Max. Heart Rate" ,xlabel="Age",ylabel="Max. Heart Rate");
ax1.legend(*sct.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=16,fontweight="bold");
```
